# **Matplotlib: Esempi e Test Interattivi**

Questo notebook contiene esempi pratici su Matplotlib con test e commenti.

Esegui ogni cella per esplorare le funzionalit√† di visualizzazione dati.

```python
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from IPython.display import display
```

## **1. Line Chart**
Grafico a linee per mostrare la relazione tra due variabili.

```python
x = np.arange(1, 6)
y = np.array([10, 20, 25, 30, 40])
plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.title("Line Chart")
plt.xlabel("Asse X")
plt.ylabel("Asse Y")
plt.show()
```

## **2. Bar Chart**
Grafico a barre per confrontare diverse categorie.

```python
x = ['A', 'B', 'C', 'D', 'E']
y = [10, 20, 30, 40, 50]
plt.bar(x, y, color='g')
plt.title("Bar Chart")
plt.xlabel("Categorie")
plt.ylabel("Valori")
plt.show()
```

## **3. Histogram**
Istogramma per mostrare la distribuzione dei dati.

```python
data = np.random.randn(1000)
plt.hist(data, bins=20, color='purple', alpha=0.7)
plt.title("Histogram")
plt.xlabel("Valori")
plt.ylabel("Frequenza")
plt.show()
```

## **4. Scatter Plot**
Grafico a dispersione per visualizzare relazioni tra due variabili.

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y, color='r')
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

## **5. Pie Chart**
Grafico a torta per mostrare proporzioni.

```python
labels = ['A', 'B', 'C', 'D']
sizes = [30, 40, 20, 10]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['red', 'blue', 'green', 'yellow'])
plt.title("Pie Chart")
plt.show()
```

## **6. 3D Plot**
Grafico tridimensionale per rappresentare dati complessi.

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
ax.plot_surface(X, Y, Z, cmap='viridis')
plt.title("3D Plot")
plt.show()
```
